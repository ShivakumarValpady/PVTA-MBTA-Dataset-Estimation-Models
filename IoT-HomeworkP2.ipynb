### 8) Refactor your solution from question 7) to create a function or set of functions that returns the hourly max passenger count for an inputed 'RouteId'.  Then plot a grouped bar chart using the result of this function for 'RouteId's [ '30043', '20031',  '10029'] .


The historical dataset used for this question is the same as in question 6 and 7.

Plot.ly example for a grouped bar chart:  https://plot.ly/python/bar-charts/#grouped-bar-chart

```python
with open('D:/First Sem/IOT/Programming Assignments/IOT-Homework-P2/pvta_08_2018.json')as file2:
    data7 = json.load(file2)
   
```

```python
 df7 = pd.io.json.json_normalize(data7, record_path='Vehicles')
dropcolumn = ['BlockFareboxId','CommStatus','Destination','Deviation','DirectionLong','DisplayStatus','DriverName','GPSStatus','Heading','OpStatus','RunId','Speed']
df7.drop(dropcolumn, axis=1, inplace=True)

```

```python
df7['LastUpdated'] = df7.apply(lambda x: pd.to_datetime(x['LastUpdated'])-pd.to_timedelta(4,'h'),axis=1)
df7 = df7[(df7['LastUpdated'].dt.hour>=7) & (df7['LastUpdated'].dt.hour<=22)]
```

```python
def PassengerMax(routeid):
    dfxx = df7[df7.RouteId == routeid]
    dff7 = dfxx.groupby(dfxx['LastUpdated'].dt.hour)['OnBoard'].max()
    trace1 = go.Bar(x = dff7.index, y= dff7.values, name = routeid)
    return trace1
```

```python
t = PassengerMax(30043)
r = PassengerMax(20031)
z = PassengerMax(10029)

```

```python
t
```

```python
py.offline.init_notebook_mode(connected = True)
```

```python
data = [t,r,z]
layout = go.Layout(barmode = 'group')

fig = go.Figure(data=data, layout=layout)
py.offline.iplot(fig, filename = 'Maximum Passengers')
```

1) The data is obtained and unwanted columns are dropped. The 'LastUpdated' Column is converted to DateTime format and the time is subtracted by 4hrs.
2) A function is defined in which for a paricular routeId, maximum passengers are obtained and the graph is plotted for the same.
3) The grouped bar plot for the three routes are obtained by plotting the data in plotly
